# 做移動平均


以下例子，每20天的移動平均
window可以讓你開小視窗,window=20代表每20個開一個視窗

```python
move = P.rolling(window=20).mean()
```

```python
P.plot()
move.plot()
P.rolling(window=60).mean().plot()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC
```

```python
df = pd.read_csv('./第六週＋第七週作業/train.csv')
df
```

### Q1: 首先，請取出所有的Age跟Sex作為題目，和Survived做為答案並分成三個同名的numpy array。  (10分)
##### (提示： Age.shape = (891, ))

```python
age = np.array(df.Age) 
sex = np.array(df.Sex) 
survived = np.array(df.Survived)
age,sex,survived
```

## Q2:其中，由於Age有一些值是空值NaN，請把這些空值都先換成0。   (10分)
#### （提示： 可使用 np.isnan(某個你想確認是不是NaN的年齡) ）

```python
age[(np.isnan(age))] = 0
age
```

### Q3:再來，Sex部分因為都是文字，不方便我們做svm，請把Male都換成0，Female都換成1   （10分）

```python
for i,j in enumerate(sex):
    if (sex[i] == 'male'):
        sex[i] = 0
    elif (sex[i] == 'female'):
        sex[i] = 1
sex
```

## Q4:由於我們想把Sex跟Age都當成data，請把Sex和Age兩個array合併成一個。（10分）
## （提示：先用np.reshape，再用np.append）

```python
#age = np.reshape(age,(891,1))
age.shape = (891,1)
sex = np.reshape(sex,(891,1))
age,sex
#sexPlusAge = np.column_stack((sex,age))
sexPlusAge = np.c_[sex,age]
sexPlusAge
```

## Q5: 實際拿svm 訓練看看，並用同樣的data，preict看看結果（這次不用分training data跟testing data）
#### （這次是因為前面做過了，一般都還是要噢！）(10分)

```python
clf = SVC()
```

```python
survived = np.reshape(survived,(891,-1))
x = sexPlusAge
y = survived
x,y
```

```python
clf.fit(x,y)
```

```python
clf.predict([[0,30]]),clf.predict([[1,32]]),clf.predict([[0,74]])
```

```python
clf.predict(x)
```

```python
clf.score(x,y) #平均預測正確率(Returns the mean accuracy on the given test data and labels) 
```

```python
def adscore(mat1):
    a=np.zeros((3,30))
    for i in range(3):
        for j in range(30):
            final1=mat1[i,j,0]*0.2+mat1[i,j,1]*0.35+mat1[i,j,2]*0.45
            final2=mat1[i,j,1]*0.4+mat1[i,j,2]*0.6
            if (final1>=final2):
                a[i,j]=final1
            else:
                a[i,j]=final2
    
    return a
```

```python
adscore(mat1)
```
